## Main Task — Number-Guessing Game

Build a working Number-Guessing Game inside a function `play_game()`:

1. The computer picks a random number between 1 and 100.
2. Keep asking the player to guess (use a loop).
3. After each guess, print `"Too high"` or `"Too low"`.
4. Count the number of guesses.
5. When correct, congratulate the player and show the guess count.

Start from what we built in class, then make it your own.

```python
import random

# Your game here
def play_game(): 
    secret_number = random.randint(1,100)
    guess_count = 0 

    print("Welcome to the number guessing game! 👾")
    print("I have a number between 1 and 100 in my mind! ")


    while True:
        guess = int(input("Enter your guess: "))
        guess_count = guess_count + 1

        if guess > secret_number: 
            print("Too High")
        elif guess < secret_number: 
            print("Too Low")
        else: 
            print("Congrats You Guessed Correct!", secret_number,"🎇")
            print("You used ", guess_count ,"chances!")
            break
    
```

```python
play_game()
```
